```python
# Adding new column 'Name' using mimesis library

from mimesis import Generic

# Create a Mimesis instance for generating names
random_name = Generic()

# Generate random names and add them as a new column
df['Name'] = [random_name.person.full_name() for i in range(len(df))]

# display name column
df['Name'].head()
```

```python
# Adding new column 'Race' using mimesis library

from mimesis import Generic

# Create a Mimesis instance for generating random race
random_race = Generic()

# Generate random race values for the 'Race' column
df['Race'] = [random_race.random.choice(['White', 'Black', 'Asian', 'Hispanic']) for j in range(len(df))]

# display race column
df['Race'].head()
```

```python
df
```

```python
# checking for missing values

total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing.head()
```

The percentage of missing value is very small. So we can drop the rows containg missing values

```python
# drop missing value rows

df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# Handling instances with whitespaces

for k in df:
    df[k]=np.where(df[k]==' ',np.NAN,df[k])
df.isnull().sum()
```

```python
# Nan rows were dropped

df.dropna(axis = 0, inplace = True)
df.isna().sum()
```

```python
# Handling instances with double whitespaces

for k in df:
    df[k]=np.where(df[k]=='  ',np.NAN,df[k])
df.isnull().sum()
```

```python
# Nan rows were dropped

df.dropna(axis = 0, inplace = True)
df.isna().sum()
```

```python
# Histogram plot function

def plot_hist(variable):
    plt.figure(figsize=(9,3))
    plt.hist(df[variable],bins=40)
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title("{} distribution with histogram".format(variable))
    plt.show()
```

```python
numerical_var = ['EaseofUse', 'Effectiveness', 'Satisfaction' , 'UsefulCount']

for n in numerical_var:
    plot_hist(n)
```

```python
# Box plot function

def box_plot(variable):
    plt.figure(figsize=(9,3))
    plt.boxplot(df[variable])
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title("Box plot of {}".format(variable))
    plt.show()

# box plot
for n in numerical_var:
    box_plot(n)
```

```python
# created new feature sideeffects

mapping = {1:5, 2:4, 3:3, 4:2,5:1} 
# 5:'Extremely severe side effects',4:'Severe side effects',3:'Moderate side effects',2:'Mild side effects',1:'No side effects'}
df['SideEffects'] = df['Satisfaction'].map(mapping)
```

```python
# Pie chart of Side_Effects

Side_Effects = df['SideEffects'].value_counts()
label = df['SideEffects'].value_counts().index
plt.figure(figsize=(6,6))
plt.pie(Side_Effects,labels=label, autopct='%1.1f%%')
plt.legend()
plt.show()
```

30.1% has no side effects, 29.2% has extremely severe side effects
